```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
data = {
    'country': ['GB', 'GB', 'GB', 'IN', 'IN', 'IN', 'US', 'US', 'US'],
    'channel_title': [
        'The Tonight Show Starring Jimmy Fallon', 'TheEllenShow', 'Jimmy Kimmel Live',
        'VikatanTV', 'etvteluguindia', 'Flowers Comedy',
        'ESPN', 'The Tonight Show (US)', 'Vox'
    ],
    'trending_video_count': [208, 207, 207, 284, 282, 270, 203, 197, 193]
}

df = pd.DataFrame(data)

# Set custom warm color palette
warm_palette = {
    'GB': '#FF9999',   # Light coral
    'IN': '#FFCC99',   # Light orange
    'US': '#99CCFF'    # Light blue
}

# Set country order to match colors
country_order = ['GB', 'IN', 'US']

# Plot
plt.figure(figsize=(12, 8))
sns.barplot(
    data=df,
    x='trending_video_count',
    y='channel_title',
    hue='country',
    hue_order=country_order,
    palette=warm_palette
)

# Styling
plt.title('Top 3 Trending Creators by Country', fontsize=18, fontweight='bold')
plt.xlabel('Trending Video Count', fontsize=14, fontweight='bold')
plt.ylabel('Channel Title', fontsize=14, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Country', title_fontsize=13, fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Indian creators have higher trending video counts, indicating faster growth and engagement in regional content.

**Business Takeaway:**
India is a strategic growth market. Invest in local influencers, vernacular content, and culturally relevant storytelling.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data
data = {
    'title': [
        'RollerCoaster Tycoon Campaign',
        'Ajit Pai PSA (GB)',
        'Ajit Pai PSA (US)',
        'Net Neutrality Repeal',
        'LuLaRoe Consultant Controversy'
    ],
    'channel_title': [
        'RollerCoaster Tycoon',
        'Daily Caller',
        'Daily Caller',
        'Washington Post',
        'J S'
    ],
    'country': ['GB', 'GB', 'US', 'US', 'US'],
    'dislike_like_ratio': [29.05, 24.18, 23.94, 22.35, 20.71]
}

df = pd.DataFrame(data)

# Define color palette per country
country_colors = {
    'US': '#2B6CB0',   # Orange
    'GB': '#B39DDB',   # Light Blue
}

# Plotting
plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    data=df,
    x='title',
    y='dislike_like_ratio',
    hue='country',
    palette=country_colors
)

# Add text labels above bars
for i in range(len(df)):
    plt.text(i, df['dislike_like_ratio'][i] + 0.5,
             f"{df['dislike_like_ratio'][i]:.2f}",
             ha='center', fontsize=10)

# Titles and labels
plt.title('Most Disliked Videos (Dislike-to-Like Ratio)', fontsize=16, weight='bold')
plt.xlabel('Video Title', fontsize=12, weight='bold')
plt.ylabel('Dislike-to-Like Ratio', fontsize=12, weight='bold')
plt.xticks(rotation=20)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

Business Insight: Negative Sentiment Monitoring on YouTube
The visualization identifies the top 5 most disliked videos across GB and US based on the dislike-to-like ratio.

Key observations include:

Highest negativity was observed for the "RollerCoaster Tycoon Campaign" (🇬🇧 GB), indicating backlash possibly due to unmet user expectations or promotional missteps.

PSAs from Chairman Ajit Pai and the Net Neutrality Repeal videos from 🇺🇸 US also triggered substantial viewer dissatisfaction, suggesting these topics are controversial and prone to public scrutiny.

The presence of news and political content in the top disliked list reveals higher sensitivity among viewers toward regulatory or socio-political messaging.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data from screenshot
data = {
    'country': ['US', 'US', 'US', 'US', 'US', 'GB', 'GB', 'GB', 'GB', 'GB'],
    'category_name': [
        'Howto & Style', 'Howto & Style', 'Howto & Style', 'Howto & Style', 'Howto & Style',
        'Gaming', 'Music', 'People & Blogs', 'Gaming', 'Comedy'
    ],
    'comments_per_view': [
        0.1176, 0.1135, 0.1116, 0.1104, 0.1073,
        0.0731, 0.0728, 0.0602, 0.0587, 0.0577
    ]
}
df = pd.DataFrame(data)

# Group by country + category
grouped_df = df.groupby(['country', 'category_name']).mean(numeric_only=True).reset_index()

# Wine tones palette
wine_palette = [
    '#6A1B9A', '#AD1457'  # Assuming 2 countries: US and GB
]

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=grouped_df,
    x='category_name',
    y='comments_per_view',
    hue='country',
    palette=wine_palette
)

plt.title('Conversation Intensity by Category and Country', fontsize=14, fontweight='bold')
plt.xlabel('Category', fontweight='bold')
plt.ylabel('Avg. Comments per View', fontweight='bold')
plt.xticks
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

“How-To & Style” Sparks the Most Conversation
This category shows the highest comments-per-view, especially in the US.

**Business Takeaway:**
Use interactive formats (polls, comment questions, duets) when posting tutorial or style content to maximize community engagement.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'country': ['US', 'IN', 'GB'],
    'Subscribe Mentions': [18749, 21144, 13269],
    'Links to External Sites': [35369, 30565, 32731],
    'Watch Full CTA': [1090, 1357, 933],
    'Shop or Buy CTA': [424, 49, 381],
    'Comment Callouts': [256, 205, 109],
    'Smash That Like CTA': [0, 0, 0]
}

df = pd.DataFrame(data)
df = df.set_index('country').T  # Transpose to flip axes

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, fmt='g', cmap='Blues', linewidths=0.5, linecolor='gray')

# Labels and title
plt.title('Call-to-Action Mentions by Country', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontweight='bold')
plt.ylabel('Call to Action Type', fontweight='bold')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Business Insights: Call-to-Action (CTA) Patterns Across Countries
🔗 "Links to External Sites" dominate across all countries, with the US leading at 35,369 mentions. This suggests creators are heavily driving traffic off-platform—likely for monetization or promotional partnerships.

🇮🇳 India shows the highest mentions for "Subscribe" CTAs (21,144), indicating a growth-driven strategy where creators focus on building a loyal subscriber base.

🇺🇸 The US emphasizes more diverse CTA usage, including "Shop or Buy" (424) and "Comment Callouts" (256)—implying stronger audience activation strategies, encouraging interactions beyond passive viewing.

🇬🇧 The UK shows balanced engagement in linking and comment callouts but relatively lower subscription and purchase CTA mentions, suggesting a possible reliance on content virality rather than community or commercial engagement.

🛒 "Shop or Buy" CTAs are least used in India, indicating a potential opportunity for e-commerce integration growth in Indian YouTube content strategies.

❌ "Smash That Like CTA" is surprisingly unused across all countries—possibly due to changes in platform culture or shifts in content strategies favoring subtler CTAs.

💡 Business Takeaways
Optimize for Off-Platform Value: Brands and creators should prioritize “external site” CTAs to redirect traffic to affiliate sites, e-shops, or content hubs.

Target Region-Specific Behavior:

In India, focus on driving subscriptions and long-term retention.

In the US, invest in comment-driven and product-based CTAs to boost interactions and conversions.

In the UK, explore ways to convert high linking into more subscriber or product actions.

Enhance CTA Diversity: Experiment with underused CTA types like comment prompts or interactive shopping links for greater viewer engagement.
